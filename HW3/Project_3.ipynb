```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plot
X = LifeExpectancydf.loc[:, 'Life_Expectancy'].values.reshape(-1,1)
Y = LifeExpectancydf.loc[:, 'Schooling'].values.reshape(-1,1)
LRmodel = LinearRegression()
LRmodel.fit(X, Y)
Y_pred = LRmodel.predict(X)
plot.scatter(X, Y)
plot.plot(X, Y_pred, color='red')
plot.show()
```

```python
print("The slope: ", LRmodel.coef_)
print("The intercept: ", LRmodel.intercept_)
```

```python
tempDF = LifeExpectancydf.copy()

tempDF['Predicted_Score'] = Y_pred

tempDF
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("MSE: ", mean_squared_error(Y, Y_pred))
print("R2: ", r2_score(Y, Y_pred))
```
