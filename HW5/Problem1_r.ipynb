2 .Recall the dataset contains amounts spent on different types of advertisements and the resulting Sales. Build the following linear models and plot the obtained MSE and R2 values for each model

a**. Sales vs. TV**

```python
#finding slope and intercept for the line
%%R

data("company_dataDF")
x <- company_dataDF$TV
y <- company_dataDF$Sales
linear_model1 <- lm(y ~ x, company_dataDF)
linear_model1 
```

# **Linear Model:**Sales=6.97 + 0.05 * TV

```python
#Finding the R2 value 
%%R
cat('R2:',summary(linear_model1)$r.squared)
cat('\n')
cat('MSE:',mean(linear_model1 $residuals^2))
```

2.Recall the dataset contains amounts spent on different types of advertisements and the resulting Sales. Build the following linear models and plot the obtained MSE and R2 values for each model

**b.Sales vs. (TV, Radio, Newspaper)**

```python
#Finding slope and intercept for 'Sales vs. (TV, Radio, Newspaper)'
%%R

data("company_dataDF")
x <- company_dataDF$Sales
y <- company_dataDF$TV+company_dataDF$Radio+company_dataDF$Newspaper

linear_model <- lm(x ~ y, company_dataDF,)
linear_model
```

```python
#Finding the R2 value 
%%R
cat('R2:',summary(linear_model)$r.squared)
cat('\n')
cat('MSE:',mean(linear_model $residuals^2))
```

**Spliting the Data into Training and Test Cases**

```python
#Training the data
%%R
library("caret")

set.seed(140)
Index <- createDataPartition(company_dataDF$Sales, p = 0.7,
                                  list = FALSE,
                                  times = 1)
training_data <- company_dataDF[Index,]
test_data <- company_dataDF[-Index,]
```

```python
#Finding the dimensions of training data of sales attribute
# 70% of 200 cases are used for training
%%R

dim(training_data['Sales'])
```

```python
#Finding the dimensions of test data of sales attribute
# 30% of 200 cases are used for testing
%%R

dim(test_data['Sales'])
```

**Linear Model of Sales Vs TV**

```python
#finding slope and intercept for the line
%%R

data("company_dataDF")
linear_model2 <- lm(Sales ~ TV, data=training_data,)
linear_model2 
```

```python
#Finding the R2 value and MSE values
%%R
cat('R2:',summary(linear_model2)$r.squared)
cat('\n')
cat('MSE:',mean(linear_model2$residuals^2))
```

**Linear Model for Sales Vs (Radio,TV,NewsPaper)**

```python
%%R

Xtraining_data <- training_data$Sales
Ytraing_data <- training_data$TV+training_data$Radio+training_data$Newspaper

data("company_dataDF")
trained_linearModel <- lm(Xtraining_data ~ Ytraing_data, data=training_data,)
trained_linearModel
```

```python
#Finding the R2 value and MSE values
%%R
cat('R2:',summary(trained_linearModel)$r.squared)
cat('\n')
cat('MSE:',mean(trained_linearModel$residuals^2))
```

**Summary:**

*   Model trained and tested on the entire data set of 200 cases for Sales Vs TV:

       > Sales=6.97 + 0.05 * TV

*   Model trained on 70% of the data and tested on 30% of the data for Sales Vs TV:
       
       > Sales = 6.83 + 0.05 * TV

*   Model trained and tested on the entire data set of 200 cases for Sales Vs (all=TV,Radio,Newspaper):
 
       > Sales = 4.57 + 0.05 * all

*   Model trained on 70% of the data and tested on 30% of the data for Sales Vs (all=TV,Radio,Newspaper):

      > Sales = 4.91 + 0.05 * all



```python
#Plotting for best fit linear model
%%R
plot(Sales ~ TV,data = company_dataDF, pch = 19,col="cyan",xlab = "TV", ylab = "Sales")
abline(linear_model1, col="red")
abline(linear_model2, col="blue")
legend("topleft", legend=c("Model trained and tested on 200 examples (R2=0.8121)", "Model trained on 70% and tested on 30% of data (R2=0.8400)"), col=c("red", "green"), lty=1:2, cex=0.8)
```

```python
%%R
plot(Sales ~ TV,data = company_dataDF, pch = 19,col="cyan",xlab = "TV", ylab = "Sales")
abline(linear_model, col="red")
abline(trained_linearModel, col="darkmagenta")
legend("topleft", legend=c("Model trained and tested on 200 examples (R2=0.8554)", "Model trained on 70% and tested on 30% of data (R2=0.8704)"), col=c("red", "green"), lty=1:2, cex=0.8)
```
