```python
#importing the data from Git

import pandas as pd

adult_trainingDF = pd.read_csv("https://raw.githubusercontent.com/MugdhaYada/my_projects-IDS/main/HW4/adult_training.csv");

adult_trainingDF
```

I have given the path of my git repo where I have stored the data in the form of CSV.

```python
#Contingency table for workclass and sex

p_crosstab = pd.crosstab(adult_trainingDF.workclass, adult_trainingDF.sex)

print(p_crosstab)
```

Created the contingency table using 'crosstab' keyword for 'Work class' & 'Sex'.

```python
# Plotting barplot of workclass and sex

barplot = p_crosstab.plot.bar(rot=20)
```

Bar plotting the contingency table of Work 'class' & 'Sex'

```python
#Contingency table for sex and marital-status

p_crosstab = pd.crosstab(adult_trainingDF.sex, adult_trainingDF.marital_status)

print(p_crosstab)
```

Created the contingency table using 'crosstab' function for 'Sex' & 'marital_status'.

```python
# Plotting barplot of sex and marital_status

barplot = p_crosstab.plot.bar(rot=20)
```

Bar plotting the contingency table of Work 'Sex' & 'marital_status'

```python
#computing the basic stats of education field to find out the outliers
adult_trainingDF['education'].describe()
```

Calculating the count, min, max, std, and the quartiles of the education field using 'describe'.

```python
#Determining the outliers using the three sigma rule for education attribute
import matplotlib.pyplot as plt

adult_trainingDF_outliers = adult_trainingDF.query('(education > 3 | education < -3)')['education']

adult_trainingDF_outliers.plot.density()
plt.title('Density Plot for Education Outliers')
plt.show()
```

Finding the outliers of the education field using the three sigma rule.
The z-values that are either greater than 3, or less -3 are outliers.

```python
#create a box plot for outliers
import plotly.express as px

boxplot = px.box(adult_trainingDF, y="education")

boxplot.show()
```

```python
#Standardizing the age by calculating z-score
from scipy import stats
import matplotlib.pyplot as plt


adult_trainingDF['scaled_age'] = stats.zscore(adult_trainingDF.age, nan_policy='omit')

adult_trainingDF.age.plot.density(color="cyan")
plt.title('Density plot for Unscaled Age')
plt.show()


adult_trainingDF.scaled_age.plot.density(color="black")
plt.title('Density plot for Scaled Age')
plt.show()
```

Standardizing the age variable by calculating the z-score and then creating the density plot for age.

```python
#computing the basic stats of age field to find out the outliers
adult_trainingDF['scaled_age'].describe()
```

Calculating the count, min, max, std, and the quartiles of the age field using 'describe'.

```python
#Determining the outliers using the three sigma rule for age attribute

import matplotlib.pyplot as plt

adult_trainingDF_ageOutliers = adult_trainingDF.query('(scaled_age > 3 | scaled_age < -3)')['scaled_age']
adult_trainingDF_ageOutliers.plot.density()
plt.title('Density Plot for age Outliers')
plt.show()
```

Determining the outliers for 'age' attribute using the three sigma rule.

```python
#create a box plot
import plotly.express as px

boxplot = px.box(adult_trainingDF, y="age")

boxplot.show()
```

Above is the box plot for age field and the dotted points are outliers.
by looking at the outliers pointed in boxplot we can say that,The most extreme outlier is at age 90.
